# 01. Post method

```python
url = "https://6009686077055686.6.gcp.databricks.com/api/2.1/jobs/run-now"

payload = {
  "job_id": 545929794148080,
  "notebook_params": {
    "sections": "Job description|Job Benefits|Job Requirements|Schedule",
    "vendor": "indeed.com",
    "driver_type": "Professional CDL-B for reefer",
    "salary": "125,000 per year",
    "availability": "inmediatly",
    "location": "New York"
  }
}
headers = {
  'Content-Type': 'application/json',
  'Authorization': 'Bearer ' + token
}

response = requests.request('post', url, headers=headers, data=json.dumps(payload))

json_response_content = json.loads(response.content)

json_response_content
```
